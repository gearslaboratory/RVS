```python
import rasterio
import matplotlib.pyplot as plt
```

```python
raster_file = f'C:/Users/ingli/OneDrive - University of Nevada, Reno/Desktop/unr/git/RVS/240226_test/outs/rasters/biomass_year7_fire.tif'

# Open the raster file
with rasterio.open(raster_file) as src:
    # Read the raster data
    raster_data = src.read(1)  # 1 means we read the first band


# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 10))

# Display the raster data
cmap = plt.cm.viridis  # Choose a colormap
cax = ax.imshow(raster_data, cmap=cmap, interpolation='nearest')

# Add a color bar
fig.colorbar(cax, ax=ax, label='Total Biomass')

# Add labels and title if needed
ax.set_title('Total Biomass Raster Plot')
ax.set_xlabel('Longitude Index')
ax.set_ylabel('Latitude Index')

# Show the plot
plt.show()
```
